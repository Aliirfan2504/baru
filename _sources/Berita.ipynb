```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
kompas=requests.get('https://www.kompas.com/')
kompas.content
berita_list=[]
```

```python
beautify=BeautifulSoup(kompas.content)
```

```python
beautify.find_all('div',{'class','col-bs10-7'})
```

```python
beautify.find_all('div',{'class','latest ga--latest mt2 clearfix'})
```

```python
beautify.find_all('div',{'class','article__list clearfix'})
```

```python
wadahjenis = beautify.find_all('div',{'class','article__subtitle article__subtitle--inline'})
```

```python
wadahdate = beautify.find_all('div',{'class','article__date'})
```

```python
wadahj = beautify.find_all('h3',{'class','article__title article__title--medium'})
```

```python
wadahl = beautify.find_all('a',{'class','article__link'})
```

```python
for i in wadahj:
    judul=i.find('a',{'class','article__link'}).text
    link=i.a.get('href')
    print(judul)
    print(link)
    berita_list.append([judul,link])
```

```python
berita=pd.DataFrame(berita_list,columns=["judul","link"])
```
